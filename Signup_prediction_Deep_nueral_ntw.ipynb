# Regression

## Import Data

```python
import pandas as pd 
import dataiku
from dataiku import pandasutils as pdu

mydataset = dataiku.Dataset("PA_reg3")
df = mydataset.get_dataframe()
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

## Data Preprocessing

```python
df['Month'].replace({
        'January': 1,
        'February': 2,
        'March': 3,
        'April':4,
         'May':5,
         'June':6,
         'July':7,
         'August':8,
         'September':9,
         'October':10,
         'November':11,
         'December':12}
        , inplace=True)
```

```python
df.head()
```

```python
df_dum=pd.get_dummies(df)

df_dum['signuprate']=df['SignupRate']
df_dum.drop(columns='SignupRate',inplace=True)
df_dum.head(5)
```

```python
df_dum.columns = df_dum.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df_dum.columns = df_dum.columns.str.strip().str.lower().str.replace('&', '')
```

```python
df_dum.columns = df_dum.columns.str.strip().str.lower().str.replace('/', '_')
```

```python
df_dum.columns = df_dum.columns.str.strip().str.lower().str.replace('Ã©', 'e')
```

```python
df_dum.columns = df_dum.columns.str.strip().str.lower().str.replace("'", '')
```

```python
df_dum.head()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

## train test split

```python
y_val= df_dum["signuprate"]
x_data=df_dum.drop("signuprate",axis=1)
```

```python
X_train, X_eval,y_train,y_eval=train_test_split(x_data,y_val,test_size=0.3,random_state=101)
```

## Data Transformation

```python
scaler_model = MinMaxScaler()
scaler_model.fit(X_train)

X_train=pd.DataFrame(scaler_model.transform(X_train),columns=X_train.columns,index=X_train.index)

scaler_model.fit(X_eval)

X_eval=pd.DataFrame(scaler_model.transform(X_eval),columns=X_eval.columns,index=X_eval.index)
```

## Feature Scaling

```python
feat_cols=[]
for cols in df_dum.columns[:-1]:
    column=tf.feature_column.numeric_column(cols)
    feat_cols.append(column)
    
print(feat_cols)
```

##Building model

```python
#The estimator model
model=tf.estimator.DNNRegressor(hidden_units=[1024, 512, 256],feature_columns=feat_cols)

#the input function
input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(X_train,y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
#Training the model
model.train(input_fn=input_func,steps=1000)
```

##Evaluating model

```python
#Evaluating the model
train_metrics=model.evaluate(input_fn=input_func,steps=1000)
```

```python
#Now to predict values we do the following
pred_input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_eval,y=y_eval,batch_size=10,num_epochs=1,shuffle=False)
preds=model.predict(input_fn=pred_input_func)

predictions=list(preds)
final_pred=[]
for pred in predictions:
    final_pred.append(pred["predictions"])
```

## Predictions

```python
final_pred
```

##RMSE R square errors

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_eval, final_pred))**0.5
```

```python
from sklearn.metrics import r2_score
r2_score(y_eval, final_pred)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(dff, labels, test_size = 0.25,
                                                                           random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

## 100 estimators

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate model 
rf = RandomForestRegressor(n_estimators= 100, random_state=42)

```

```python
rf_new = RandomForestRegressor(n_estimators = 100, criterion = 'mse', max_depth = None, 
                               min_samples_split = 2, min_samples_leaf = 1)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
